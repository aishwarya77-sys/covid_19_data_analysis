```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational",],inplace=True,axis=1)
```

```python
covid_df.head(5)
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

```python
#ACTIVE CASES
covid_df['Active_Cases']=covid_df["Confirmed"]-(covid_df['Cured']+covid_df['Deaths'])
covid_df.tail(100)
```

```python
covid_df = covid_df.dropna()
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=["Confirmed", "Deaths", "Cured"], index="State/UnionTerritory", aggfunc=max)
statewise["Recovery Rate"]=statewise["Cured"]*100/statewise["Confirmed"]
statewise["Mortality Rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
#Top 10 active cases states
top_10_active_cases=covid_df.groupby(by="State/UnionTerritory").max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most active cases in India",size=25)
```

```python
ax=sns.barplot(data=top_10_active_cases.iloc[:10],y='Active_Cases',x='State/UnionTerritory',linewidth=2,edgecolor='red')
```

```python
top_10_active_cases=covid_df.groupby(by="State/UnionTerritory").max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'],ascending=False).reset_index().head(10)
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 states with most active cases in India",size=25)
colors = sns.color_palette("hsv", len(top_10_active_cases))
ax = sns.barplot(data=top_10_active_cases, y='Active_Cases', x='State/UnionTerritory', hue='State/UnionTerritory', dodge=False, palette=colors, linewidth=2, edgecolor='red', legend=False)

plt.xlabel("States")
plt.ylabel('Total Active Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'State/UnionTerritory', aggregate maximum values for 'Deaths' and 'Date', then sort and select top 10
top_10_deaths = covid_df.groupby(by="State/UnionTerritory").max()[['Deaths', 'Date']].sort_values(by=['Deaths'], ascending=False).reset_index().head(10)

# Create the plot
fig = plt.figure(figsize=(16, 9))
plt.title("Top 10 states with highest deaths in India", size=25)

# Create a bar plot with different colors for each bar
colors = sns.color_palette("hsv", len(top_10_deaths))

# Using hue and setting legend to False
ax = sns.barplot(data=top_10_deaths, y='Deaths', x='State/UnionTerritory', hue='State/UnionTerritory', dodge=False, palette=colors, linewidth=2, edgecolor='red', legend=False)

plt.xlabel("States")
plt.ylabel('Total Deaths')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Growth Trend
top_states = ['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh']
filtered_df = covid_df[covid_df['State/UnionTerritory'].isin(top_states)]

# Create the plot
plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=filtered_df, x='Date', y='Active_Cases', hue='State/UnionTerritory', linewidth=2)

# Set plot title and labels
plt.title("Growth Trend of Active Cases in Top 5 Affected States in India", size=16)
plt.xlabel("Date")
plt.ylabel("Active Cases")
plt.xticks(rotation=45)

# Show the plot
plt.legend(title='State')
plt.show()
```

```python
#Male vs Female Vaccination

total_male_vaccinations = vaccine_df['Male(Individuals Vaccinated)'].sum()
total_female_vaccinations = vaccine_df['Female(Individuals Vaccinated)'].sum()

# Create a pie plot
labels = ['Male', 'Female']
sizes = [total_male_vaccinations, total_female_vaccinations]
colors = ['#1f77b4', '#ff7f0e']  # Blue for Male, Orange for Female

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Male vs Female Vaccination Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#remove rows where state=India

vaccine = vaccine_df.loc[vaccine_df.State != 'India']

# Display the first few rows of the modified DataFrame to verify
vaccine.head() 
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
vaccine.head()
```

```python
#Most vaccinated states
max_vac=vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=max_vac, x='State', y='Total', hue='State', palette='viridis', dodge=False, legend=False)
plt.title('Top 5 Vaccinated States in India')
plt.xlabel('State')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.show()
```
